```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.checkers.co.za/c-2256/All-Departments"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')

soup
```

```python


product_names = soup.find_all("h3", class_ = "item-product__name")
product_price_standard = soup.find_all('span', class_="now")

price_list = [price.get_text() for price in product_price_standard]
product_list = [ item.get_text() for item in product_names]


for i in range(len(product_list)):
    print(f"{product_list[i]}: {price_list[i]}")
```
