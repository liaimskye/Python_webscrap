```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
```

```python
url = "https://www.checkers.co.za/c-2256/All-Departments?q=%3Arelevance%3AallCategories%3Afood%3AbrowseAllStoresFacetOff%3AbrowseAllStoresFacetOff&page="
page_number = 0
df = pd.DataFrame(columns = ["product", "price"])
product_and_price = []
browser = webdriver.Chrome()

while True:
    #page = requests.get(url + str(page_number))
    page = browser.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    product_names = soup.find_all("h3", class_ = "item-product__name")
    product_price_standard = soup.find_all('span', class_="now")
    price_list = [price.get_text().strip() for price in product_price_standard]
    product_list = [ item.get_text().strip() for item in product_names]
    for i in range(len(product_list)):
        product_and_price.append([product_list[i], price_list[i]])
    
    if len(product_list) > 0:
        page_number += 1

        for i in range(len(product_list)):
            product_and_price.append([product_list[i], price_list[i]])
        
        for row in product_and_price:
            length = len(df)
            df.loc[length] = row
    
    else:
        print('all pages read')
        break

df


    

    
```
